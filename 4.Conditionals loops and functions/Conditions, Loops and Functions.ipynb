## For loop

```python
for i in range(1,20):
  print("Number is", i)
```

```python
sumx=0

for i in range(1,10):
  sumx=sumx+i
  print("Cumulative sum is", sumx)
```

```python
sumx=0
for i in range(1,200):
  sumx=sumx+i
  if(sumx>100):
    break
  print("Cumulative sum is", sumx)
```

# Functions

```python
def mydistance(x1,y1,x2,y2):
    import math
    dist=math.sqrt(pow((x1-x2),2)+pow((y1-y2),2))
    return(dist)
```

```python
mydistance(0,0,2,2)
```

```python
mydistance(4,6,1,2)
```
