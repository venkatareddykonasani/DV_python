# Lab2: Accessing Arrays

# LAB : Placeholders and axis

```python
# Generating equidistance value array
eq_array = np.linspace(1,2,50)
eq_array
```

```python
#Reshape this array to 5,10
eq_array = eq_array.reshape(5,10)
eq_array
```

```python
# Generate random sample with 50 values
ran = np.random.random(50)
ran
```

```python
# reshape and multiply it with array 'a'
ran=ran.reshape(5,-1)
ran
```

```python
# Multiply random sample to equidistance array
out = eq_array * ran
out
```

```python
# get index of max element in each column
out.argmax(axis=1)
```

```python
#get min value in each row
out.min(axis=0)
```
